## Univariate Data Analysis
This preprocessing covers analysing a single column, and mostly gives insights on the data's distribution and other information.

We have two types of Univatiate EDA which are :
- Quantitative EDA (Analysis for a single column that holds numerical data)
- Categorical EDA (Analysis for a single column that holds a categorical data either in string format or number format) 

```python
## First plot is for univariate quantitative analysis which is to analyse a column that holds continous data using matplotlib
plt.hist(df['ApplicantIncome'])
plt.title("Frequency of applicants Income")
plt.xlabel("Applicant Income")
plt.ylabel("Frequency")
plt.show()
```

Insights gained from the plot are :
- knowing how skewed the data is, either positively or negatively
- Knowing the range of data with the highest frequency
- Knowing the possible maximum and minimum value of the data.
- The general distribution of the data and so on.

```python
## Using Seaborn, for a single categorical column, countplot is used in place of barplot.
## And you can see it labels it by default, an edge it has over matplotlib
sns.countplot(data=df, x='Gender')
plt.show()
```

Insights from the plot:
- modal class.
- Check if the column class is balanced or not.
- Estimate the amount of each class and so on.

## Mulivariate Data Analysis
This is a process where we analyse the trends and relationships between two or more columns.
There are three basic types that we are going to treat in this lesson :
- One categorical and one quantitative column analysis.
- Two or more categorical columns analysis.
- Two or more quantitative columns analysis

```python
##for one categorical and one numerical column, we are using boxplot from seaborn library
## X-axis been categorical and y_axis been numerical
sns.boxplot(data=df, x='Married', y='ApplicantIncome')
plt.show()
```

Insights from the plot :
- Married applicants earn the highest income.
- Estimate of the Highest income for both single and married applicants
- Estimate of the average of both their income and so on.

```python
## Use of seaborn to plot a numerical and two categorical columns.
sns.barplot(data=df,x='Gender', y='ApplicantIncome',hue='Married')
plt.show()
```

Insights :
- Checks for the gender, marital status and how they are earning.


```python
df.info()
```

```python
## Use of matplotlib to plot two numerical columns together
plt.scatter(df['CoapplicantIncome'],df['ApplicantIncome'])
plt.title("Graph of co applicant income and applicant income")
plt.xlabel("Coapplicant Income")
plt.ylabel("Applicant Income")
plt.show()
```

Insights :
- Shows how the two columns are distributed
- Shows the types of relationship between the numerical columns and so on.

```python
sns.kdeplot(data = df, x = 'LoanAmount', hue = 'Married')
plt.show()
```

Here a Kdeplot is used to show distribution comparing two columns, Loan Amount and gender, from the plot we can see that married people apply more for loan than those that are not married and they still have the larger variance though the distributions shows something close to same average

```python
sns.scatterplot(data = df, x = 'LoanAmount', y= 'ApplicantIncome', hue = 'Gender')
plt.show()
```

Scatter plot used here compares analyses three columns, two numerical and a categorical, and from this we can roughly deduce that Males have higher Income compared to females, Males apply for Higher loans than female too and males are more represented in the data than females too.
